```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('andhrapradesh.csv')
```

```python
df
```

```python
df.head()
```

```python
plt.figure(figsize=(8,7))
plt.bar(df.Date,df.Confirmed,label='Confirmed cases',color='blue')
plt.title("covid_analysis_june_july:andhra pradesh")
plt.xticks(df.Date,rotation=90)
plt.xlabel("CASES DAILY (1stjune-15july)",labelpad=30)
plt.ylabel("Number of Cases ",labelpad=10)
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.bar(df.Date,df.Confirmed,label='Confirmed cases',color='yellow')
plt.bar(df.Date,df.recovered,label='cured cases',color='red')
plt.bar(df.Date,df.Death ,color='green',label='Death')
plt.legend()
plt.title("covid_analysis_june_july:andhra pradesh")
plt.xticks(df.Date,rotation=90)
plt.xlabel("CASES DAILY (1stjune-15july)",labelpad=50)
plt.ylabel("Number of Cases ",labelpad=50)
plt.show()
```

```python
df.hist()
```

```python
plt.figure(figsize=(10, 10))


plt.plot(df.Date,df.Confirmed,color="pink",label='Confirmed Cases')
plt.plot(df.Date,df.recovered,color="navy",label='Cured Cases')
plt.plot(df.Date,df.Death, color="orange",label='Death Cases')
plt.legend()
#plt.bar(df_filtered.Date,df_hosp.Beds,color="grey",label='Total Samples')
plt.bar(df.Date,df.Confirmed,label='Confirmed Cases',color='skyblue')
plt.bar(df.Date,df.recovered,color="palegreen",label='Cured Cases')
plt.bar(df.Date,df.Death,color="purple",label='Death Cases')
plt.title("COVID-19 STATE ANALYSIS:ANDHRA PRADESH")
plt.xticks(df.Date,rotation=90)
plt.xlabel("Date",labelpad=100)
plt.ylabel("Number of Cases",labelpad=100)
plt.show()
```

```python
from matplotlib import style
 
style.use('ggplot')
plt.figure(figsize=(10,4))
plt.plot(df.Date,df.Newcases,label='New Cases',color='red',linewidth=5)
plt.plot(df.Date,df.Newdeaths,color='green',label='Death',linewidth=5)
plt.plot(df.Date,df.Newrecovered,label='Recovered',color='orange')

plt.legend()
plt.title("covid_analysis ANDHRA PRADESH")

plt.xticks(rotation=90)
plt.xlabel("Daily cases",labelpad=20)
plt.ylabel("Number of cases",labelpad=5)
plt.grid(True,color='k')

plt.show()
```

```python
df.plot(y=['Newcases','Newrecovered','Newdeaths'],x="Date",rot=90)
```
